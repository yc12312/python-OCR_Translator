```python
from pyscreenshot import grab
import tkinter
import win32gui
try:
    from PIL import Image
except ImportError:
    import Image
import pytesseract
from googletrans import Translator
import time

#making translator
translator = Translator()

#Globals
first = True
img = None
x1 = 0
x2 = 0
y1 = 0
y2 = 0
cap_lang = None
transFromlang = None
transTolang = None
last_result = None
now_result = None

#Get Cordinate
def get_cor():
    global x1,x2,y1,y2
    x1 = captureBoard.winfo_rootx()
    x2 = x1 + captureBoard.winfo_width()
    y1 = captureBoard.winfo_rooty()
    y2 = y1 + captureBoard.winfo_height()

    #Create Screenshot
def capture():    
    get_cor()    
    global img
    img = grab(bbox=(x1,y1,x2,y2))

    #OCR
def ocr():
    capture()
    #set language
    global cap_lang 
    cap_lang = 'jpn'
    return pytesseract.image_to_string(img , lang = cap_lang)   

def captureBoard():
    captureBoard = tkinter.Toplevel()
    captureBoard.title('CaptureBoard')
    captureBoard.geometry("320x200+100+100")
    captureBoard.resizable(True, True)
    captureBoard.wm_attributes('-alpha', 0.3) 
    
class printBoard(tkinter.Tk):
    def __init__(self):
        tkinter.Tk.__init__(self)
        self.label_text = 'None'
        self.captured_label = tkinter.Label(self, text=self.label_text)
        self.captured_label.pack()
    
    #Translation
    def trans(self, captured):
        global transFromlang 
        transFromlang= 'ja'
        global transTolang
        transTolang= 'ko'
        
        tr_results = translator.translate(captured, src= transFromlang, dest=transTolang)
        #self.captured_label.config(text=str(tr_results.text))
        self.label_text = '1'
        print(tr_results)
        
        self.after(3000,self.ocrTimer)
        
        
    def ocrTimer(self):
        global now_result
        global last_result
        global first
        
        #repeated task here
        if first:
            now_result = 'None'
            first = False
        else :
            now_result = ocr()
            print('1')
            
        if now_result != last_result :
            last_result = now_result
            if now_result != None:
                printBoard.trans(now_result)
            else : 
                self.after(5000,self.ocrTimer)     

printBoard = printBoard()
printBoard.title('printBoard')
printBoard.geometry("640x400+100+100")
printBoard.resizable(True, True)

capture = captureBoard()

printBoard.ocrTimer()
printBoard.mainloop()
```

```python
import tkinter
import tkinter.ttk
import win32gui
from pyscreenshot import grab
try:
    from PIL import Image
except ImportError:
    import Image
    
import pytesseract
from googletrans import Translator

import time

import threading
from queue import Queue

import os

#Globals
img = None
x1 = 0
x2 = 0
y1 = 0
y2 = 0

transFromlang = None
transTolang = None

cap_lang = None

original_result = 'None'
tr_results = 'None'
org_lb = None
tr_lb = None

now_result = None
past_result = None

#OCR
def ocr(q):
    while True:
        #Get Cordinate
        global x1,x2,y1,y2
        x1 = captureBoard.winfo_rootx()
        x2 = x1 + captureBoard.winfo_width()
        y1 = captureBoard.winfo_rooty()
        y2 = y1 + captureBoard.winfo_height()
    
        #Create Screenshot
        global img
        img = grab(bbox=(x1,y1,x2,y2))
    
        #set language
        global cap_lang 
        cap_lang = 'jpn'
    
        global now_result
        global past_result
        
        global tr_results
        now_result = pytesseract.image_to_string(img , lang = cap_lang)
        now_result = now_result.replace('\n', ' ')
        if(past_result != now_result): 
            past_result = now_result
            
            evt = threading.Event()
            q.put((now_result,evt))
            
            tr_results = 'waitng for tanslation!\n'
            configure(tr_lb,tr_results)
            evt.wait() 
      
         
#Translation
def trans(q):
    #making translator
    translator = Translator()

    global transFromlang 
    transFromlang= 'en'
    global transTolang
    transTolang= 'ko'
    
    global original_result
    global tr_results
    
    while True:
        data,evt = q.get()
    
        try:
            original_result = data + '\n'
            temp = translator.translate(data, src= transFromlang, dest=transTolang)
            tr_results = temp.text + '\n'
            
        except:
            original_result = 'Data is none\n'
            tr_results = 'Data is none\n'
            pass
        
        configure(org_lb,original_result)
        configure(tr_lb,tr_results)
        evt.set()
        q.task_done()

def threadStart():
    q = Queue()
    t1 = threading.Thread(target=ocr,args=(q,))
    t2 = threading.Thread(target=trans,args=(q,))
    t1.start()
    t2.start()
    q.join()

def configure(target, data):
    target.configure(text = data)
    
def printBoard(self):
    printBoard = tkinter.Toplevel()
    printBoard.wm_geometry("640x400+200+200")
    printBoard.title('printBoard')
    
    #frames
    top_frame = tkinter.Frame(printBoard, relief="solid", bd=1)
    top_frame.pack(side='top', fill='both', expand = 'true')
    
    bottom_frame = tkinter.Frame(printBoard, relief="solid", bd=1)
    bottom_frame.pack(side='bottom', fill='both', expand = 'true')
    
    #combobox
    cap_lang_val = ('chi_sim','eng','kor','jpn')
    cbx = tkinter.ttk.Combobox(top_frame,values = cap_lang_val)
    #cbx.pack(side = 'left')
    cbx.set('Image Language')
    cbx.grid(row = 0, column = 0, sticky = 'W')

    lang_val = ('zh-cn ','en','ko','ja')
    cbx1 = tkinter.ttk.Combobox(top_frame,values = lang_val)
    #cbx1.pack(side = 'left')
    cbx1.set('From')
    cbx1.grid(row = 0, column = 1, sticky = 'W')
    
    lang_val = ('zh-cn ','en','ko','ja')
    cbx2 = tkinter.ttk.Combobox(top_frame,values = lang_val)
    #cbx2.pack(side = 'top')
    cbx2.set('To')
    cbx2.grid(row = 0, column = 2, sticky = 'W')

    #labels
    global org_lb,tr_lb
    orgLabel = tkinter.Label(top_frame, text = original_result , font=("Helvetica", 10), wraplength = 500)
    #orgLabel.pack(side = 'bottom')
    org_lb = orgLabel
    orgLabel.grid(row = 1, column = 3, sticky = 'W')
    
    
    transLabel = tkinter.Label(bottom_frame, text= tr_results , font=("Helvetica", 10), wraplength = 500)
    #transLabel.pack(side='top')
    tr_lb = transLabel
    transLabel.grid(row = 0, column = 0, sticky = 'W')
    
class captureBoard(tkinter.Tk):
    def __init__(self):
        tkinter.Tk.__init__(self)

captureBoard = captureBoard() 
captureBoard.title('captureBoard')
captureBoard.geometry("320x200+100+100")
captureBoard.resizable(True, True)
captureBoard.wm_attributes('-alpha', 0.3)

printBoard(captureBoard)
threadStart()
captureBoard.mainloop()
```
